```python
#Answer-2
# A scatter plot is a visual representation of how two variables relate to each other.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(3)
```

```python
x = np.random.normal(0,2,50)
y = np.random.normal(0,2,len(x))
plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("this is scatter plot")
plt.show()
```

```python
#Answer-3
#it's provides a way to plot multiple plots on a single figur
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
ax1 = plt.subplot()
ax1.plot(x1,y1)
ax1.plot(x2,y2)
ax1.plot(x3,y3)
ax1.plot(x4,y4)
```

```python
#Answer-4
# Bar plot is a data visualization used to represent data into rectangular form.
#Bar charts usually present categorical variables, discrete variables or continuous variables grouped in class intervals
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple","Microsoft","Google","AMD"])
profit = np.array([3000,8000,1000,10000])
plt.bar(company,profit,color="g")
plt.xlabel("company name")
plt.ylabel("profit")
plt.show()
```

```python
#Answer-5
# A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum.

# Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.
box1 = np.random.normal(100,10,200)
box2 = np.random.normal(90,20,200)
plt.boxplot([box1,box2])
plt.show()
```
